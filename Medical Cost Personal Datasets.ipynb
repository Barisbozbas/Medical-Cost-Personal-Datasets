```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression,Ridge,Lasso,ElasticNet
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
from sklearn.metrics import accuracy_score
```

```python
data=pd.read_csv('insurance.csv')
```

```python
age_statistic=data['age'].value_counts()
bmi_statistic=data['bmi'].value_counts()
children_statistic=data['children'].value_counts()
sex_statistic=data['sex'].value_counts()
smoker_statistic=data['smoker'].value_counts()
region_statistic=data['region'].value_counts()
```

```python
age=data['age']
bmi=data['bmi']
children=data['children']
charges=data['charges']
sex_types=pd.get_dummies(data.sex,prefix='sex')
smoker_types=pd.get_dummies(data.smoker,prefix='smoker')
region_types=pd.get_dummies(data.region,prefix='region')
data=pd.concat([age,bmi,children,charges,sex_types,smoker_types,region_types],axis=1)
```

```python
data
```

```python
data.drop(['sex_female','smoker_no'],axis=1,inplace=True)
```

```python
data.describe()
```

```python
y=data['charges']
data.drop(['charges'],axis=1,inplace=True)
```

```python
X=data
```

```python
print(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=46)
```

```python
#Veri setinde normalizasyon işleminin yapılması
```

```python
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
X_test=scaler.fit_transform(X_test)
```

```python
#Makina Öğrenmesi Algoritmaları
```

```python
#Decision Tree Algoritması
tree_regression=DecisionTreeRegressor(random_state=42,max_depth=2)
tree_regression=tree_regression.fit(X_train,y_train)
tahmin_tree_regression=tree_regression.predict(X_test)
```

```python
#Random Forest Regression
```

```python
random_regression=RandomForestRegressor(n_estimators=100,max_depth=4,random_state=42)
random_regression.fit(X_train,y_train)
random_regression_tahmin=random_regression.predict(X_test)
```

### Lasso Regresyon Algoritması

```python
lassoReg=Lasso(alpha=2)
lassoReg.fit(X_train,y_train)
tahmin_lasso=lassoReg.predict(X_test)
```

### Elastic Regresyon Algoritması

```python
elastic_reg=ElasticNet(random_state=0)
elastic_reg.fit(X_train,y_train)
tahmin_elastic_reg=elastic_reg.predict(X_test)
```

### Ridge Regresyon Algoritması

```python
ridge_reg=Ridge()
ridge_reg.fit(X_train,y_train)
tahmin_ridge=ridge_reg.predict(X_test)
```

```python
### Sonuclar
```

```python
predicts=[tahmin_tree_regression,random_regression_tahmin,tahmin_lasso,tahmin_elastic_reg,tahmin_ridge]
```

```python
algoritma_names=["Karar Ağacı Regresyonu","Random Forrest Regresyon","Lasso Regresyon","Elastic Regresyon","Ridge Regresyon"]
```

```python
#Sonucları Hesaplama algoritması
def performance_calculate(predict):
    mae=mean_absolute_error(y_test,predict)
    mse=mean_squared_error(y_test,predict)
    rmse=np.sqrt(mse)
    r2=r2_score(y_test,predict)
    data=[mae,mse,rmse,r2]
    return data
```

```python
seriler=[]
metrics=["Mean Absolute Error (MAE)","Mean Squared Error(MSE)","Root Mean Squared Error(RMSE)","R2"]

for i in predicts:
    data=performance_calculate(i)
    seriler.append(data)
 
df=pd.DataFrame(data=seriler,index=algoritma_names,columns=metrics)
pd.set_option('display.colheader_justify','center')
```
